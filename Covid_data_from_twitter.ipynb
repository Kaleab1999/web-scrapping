```python
import twint
```

```python


c = twint.Config()

c.Search = ['covid']       # topic
c.Limit = 5000     # number of Tweets to scrape
c.Store_csv = True       # store tweets in a csv file
c.Output = "taylor_swift_tweets.csv"     # path to csv file

twint.run.Search(c)
```
