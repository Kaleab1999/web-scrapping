importting Autoscraper

```python
from autoscraper import AutoScraper
```

defining scrapping elements (url and wanted list)

```python
url='https://www.flipkart.com/search?q=samsung+m12&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_7_8_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_7_8_na_na_na&as-pos=7&as-type=RECENT&suggestionId=samsung+m12&requestId=04ac6375-9459-4539-9343-8677190c824f&as-searchtext=samsung%20'
wanted=['SAMSUNG Galaxy M12 (Blue, 64 GB)','₹10,539']
```

initiating autoscraper

```python
scraper=AutoScraper()
result=scraper.build(url,wanted)
print(result)
```

to assign label for the scrapped data

```python
scraper.get_result_similar("https://www.flipkart.com/search?q=samsung+m12&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_7_8_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_7_8_na_na_na&as-pos=7&as-type=RECENT&suggestionId=samsung+m12&requestId=04ac6375-9459-4539-9343-8677190c824f&as-searchtext=samsung%20",grouped=True)
```

label of scraped data and save the model

```python
scraper.set_rule_aliases({'rule_mry5': 'price', 'rule_7ga6':'product name'})
scraper.keep_rules(['rule_mry5','rule_7ga6'])
scraper.save('flipkart_data')
```

loading the model

```python
scraper.load('flipkart_data')
```

appling the model for other website

```python
result2=scraper.get_result_similar('https://www.flipkart.com/search?q=samsung+m11&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_3_10_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_3_10_na_na_na&as-pos=3&as-type=RECENT&suggestionId=samsung+m11&requestId=bd854cce-9c3e-40e6-9f81-359126f857b1&as-searchtext=samsung%20m1',group_by_alias=True)
```

to remove unwanted symbol '₹' inorder to change to Ethiopia Currency (Birr)

```python
new['price']=new['price'].replace({'₹':'',',':''},regex=True)
```

to change datatype of price column

```python
new.info()
```

```python
new['price']=new['price'].astype(float)
```

changing the currency

```python
new['price']=new['price'].mul(0.67)
```

```python
new.rename(columns={'price':'price in ETB'})
```
